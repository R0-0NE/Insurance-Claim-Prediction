```python
plt.pie(df["insuranceclaim"].value_counts(),labels=["Claim","No Claim"],autopct='%1.1f%%',
        pctdistance=1.25,labeldistance=1.1,hatch=['**O','.||.'])
plt.grid()
plt.show()
```

```python
corr=df.corr()
plt.figure(figsize=(15,10))
sns.heatmap(corr,annot=True,cmap="rocket_r")
plt.show()
```
